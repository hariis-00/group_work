```python
import pymongo
import os

# Holen Sie sich die MongoDB URI aus der Umgebungsvariable
mongo_uri = os.getenv('MONGO_URI', 'mongodb://localhost:27017/appearances')
client = pymongo.MongoClient(mongo_uri)

# Verbinden Sie sich mit der Datenbank
db = client.appearances

# Verbinden Sie sich mit der Sammlung
collection = db.appearances

# Beispielabfrage: Alle Dokumente in der Sammlung finden
documents = collection.find()
for doc in documents:
    print(doc)
```
